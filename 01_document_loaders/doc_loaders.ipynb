## LangChain Document Loaders

### CSV Loader

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(file_path="../sample_data/sample_csv_data.csv")

csv_data = loader.load()
```

```python
print(csv_data[0])
```

### Text Loader

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader(file_path="../sample_data/sample_text_data.txt")

text_data = loader.load()
```

```python
print(text_data[0])
```

### PDF Loader

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader(file_path="../sample_data/sample_pdf_data.pdf")

pdf_data = loader.load()
```

```python
print(pdf_data[0])
```

### JSON Loader

```python
from langchain_community.document_loaders import JSONLoader

loader = JSONLoader(
    file_path="../sample_data/sample_JSON_data.json",
    jq_schema=".sales_records[]",
    text_content=False
)
json_data = loader.load()
```

```python
print(json_data[0])
```
