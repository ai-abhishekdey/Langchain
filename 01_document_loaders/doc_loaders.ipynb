## LangChain Document Loaders

### CSV Loader

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(file_path="data/sample_csv_data.csv")

csv_data = loader.load()
```

```python
print(csv_data)
```

### Text Loader

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader(file_path="data/sample_text_data.txt")

text_data = loader.load()
```

```python
print(text_data)
```

### PDF Loader

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader(file_path="data/sample_pdf_data.pdf")

pdf_data = loader.load()
```

```python
print(pdf_data)
```
