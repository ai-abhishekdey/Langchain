### Load HuggingFace all-MiniLM-L6-v2 embedding model

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model="sentence-transformers/all-MiniLM-L6-v2")
```

### Get Text embeddings of a single query

```python
query="New Delhi is the capital of India"

embedding_vector=embeddings.embed_query(query)
```

```python
print(embedding_vector)
```

```python
print("Dimension of embedding vector : ", len(embedding_vector))
```

### Documents

```python
documents=[
    "Kolkata is the capital of West Bengal",
    "Dispur is the capital of Assam",
    "Mumbai is the capital of Maharastra",
    "Chennai is the capital of Tamil Nadu",
    "Kohima is the capital of Nagaland"
]
```

### Get Text embeddings of a documents

```python
embedding_vector_docs=embeddings.embed_documents(documents)
```

```python
print(embedding_vector_docs)
```

```python
print("No. of embedding vectors : ", len(embedding_vector_docs))
print("Dimension of each embedding vector : ", len(embedding_vector_docs[0]))
```
